```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt

# Import NY FosterCare data files
DataPath = "Data/NY_FosterCare.csv"

# Read the data
NY_FosterCare = pd.read_csv(DataPath)

# Display the data table for preview
NY_FosterCare
```

```python
# Create a table that shows the total number of admission and discharges by year (in all counites)
Annual = NY_FosterCare.groupby("Year").agg(
    sAdmissions=pd.NamedAgg(column="Admissions", aggfunc="sum"),
    sDischarges=pd.NamedAgg(column="Discharges", aggfunc="sum"))

# Display the new table
Annual
```

```python
# Plotting the results
Annual.plot()
```
